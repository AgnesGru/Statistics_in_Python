```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
noworodki = pd.read_csv("C:/Users/User/GIT/Statystyka w Python/Statistics_in_Python/arkusze_do_analiz/LowBirthWeight.csv", sep = ";")
noworodki.head()
```

### Countplot

```python
sns.countplot(x = "Niska waga dziecka", data = noworodki)
```

### Barplot

```python
sns.barplot(x = "Nadcisnienie", y = "Waga matki", data = noworodki)
```

### Boxplot

```python
sns.boxplot(x = "Niska waga dziecka", y = "Waga matki", data = noworodki, palette = "rainbow")
```

### stripplot

```python
sns.stripplot(x = "Wiek", y = "Waga matki", data = noworodki)
```

```python
sns.stripplot(x = "Nadcisnienie", y = "Waga matki", data = noworodki)
```

### swarmplot

```python
sns.swarmplot(x = "Nadcisnienie", y = "Waga matki", hue = "Palenie", data = noworodki)
```

### heatmap (correlation)

```python
df_waga_corr = noworodki[["Waga matki", "Wiek"]]
sns.heatmap(df_waga_corr.corr(), cmap = "coolwarm", annot = True)
```

### distribution of dataset

```python
sns.displot(noworodki["Waga matki"], kde = True)
```

```python
sns.displot(noworodki["Waga matki"], bins = 10)
```

### pairplot

```python
sns.pairplot(noworodki, hue = "Palenie")
```

### regression

```python
sns.lmplot(x = "Wiek", y = "Waga matki", data = noworodki)
```
